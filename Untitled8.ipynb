phone number

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('/','')
df["Phone_Number"] = df["Phone_Number"].str.replace('|','')
df["Phone_Number"] = df["Phone_Number"].str.replace('-','')


df
#เอาเครื่องหมายที่ไม่ต้องการออก
```

```python
df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
#ไม่ต้องมีอันนี้ก็ได้
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) #เปลี่นให้เป็นชนิดเดียวกัน
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
#ใส่เครื่องหมายให้เหมือนกัน
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
#ตัดทิ้ง
```

```python
split_address = df['Address'].str.split(', ', expand=True)
split_address
```

```python

split_columns = df['Address'].str.split(', ', expand=True)
```

```python
df[['Street_Address', 'City', 'State']] = split_columns
```

```python
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
df
#หรือแบบที่ตัวอยางให้เลยแต่ว่าตัดตัวเลขบางอย่างออก
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')

df
```

```python
#df = df.replace('N/a','')
#df = df.replace('NaN','')


df=df.fillna('')
df
```

```python
df = df.replace('N/a','')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
#ทีเป็นYจะหายไป
```
